```python
import pandas as pd

df = pd.read_csv('healthcare_dataset.csv')

df.head()
```

```python
df.info()
```

```python
df = df.drop(columns=['Name'])

df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])

df.info()
```

```python
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Medical Condition')
plt.title('Distribution of Medical Conditions')
plt.xlabel('Number of Patients')
plt.ylabel('Medical Condition')
plt.show()
```

```python
df['Medical Condition'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Gender')
plt.title('Gender Distribution of Patients')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.show()
```

```python
df['Gender'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```

```python
df['Age'].value_counts().sort_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.countplot(data=df, y='Medical Condition', hue='Gender')
plt.title('Distribution of Medical Conditions by Gender')
plt.xlabel('Number of Patients')
plt.ylabel('Medical Condition')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.countplot(data=df, y='Medical Condition', hue='Blood Type')
plt.title('Distribution of Medical Conditions by Blood Type')
plt.xlabel('Number of Patients')
plt.ylabel('Medical Condition')
plt.legend(title='Blood Type')
plt.show()
```

```python
df['Admission Year'] = df['Date of Admission'].dt.year

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

yearly_counts = df.groupby(['Admission Year', 'Medical Condition']).size().reset_index(name='Number of Patients')

plt.figure(figsize=(14, 8))
sns.lineplot(data=yearly_counts, x='Admission Year', y='Number of Patients', hue='Medical Condition', marker='o')

plt.title('Trend of Medical Conditions Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Patients Admitted')
plt.grid(True)
plt.show()
```
